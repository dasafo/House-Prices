## Outliers

Vamos a empezar viendo la relación entre el precio del piso y el área de superficie de la parcela.

```python
fig, ax = plt.subplots()
ax.scatter(x = train['GrLivArea'], y = train['SalePrice'])
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
```

Vemos `2`outliers abajo a la derecha que vamos a quitar.

```python
#Deleting outliers
train = train.drop(train[(train['GrLivArea']>4000) & (train['SalePrice']<300000)].index)
```

```python
#Check the graphic again
fig, ax = plt.subplots()
ax.scatter(train['GrLivArea'], train['SalePrice'])
plt.ylabel('SalePrice', fontsize=13)
plt.xlabel('GrLivArea', fontsize=13)
plt.show()
```

## Análisis de SalePrice

```python
#histogram
sns.distplot(train['SalePrice'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

* Vemos que la gráfica superior no se ajusta a una normal y tiene un sesgo positivo.
* Viendo la gráfica inferior podemos ver que no sigue una relación lineal lo que nos indicaría que el residuo de sus errores no siguen una distribución normal.

Sabemos según la teoría que para distribuciones con un sesgo positivo podemos aplicarle una transformación logarítmica y obtendremos una relación lineal de los residuos de los errores:

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
train["SalePrice"] = np.log1p(train["SalePrice"])
```

```python
#Check the new distribution 
sns.distplot(train['SalePrice'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

Vemos que aplicando la transformación logaritmica tenemos una distribución normal.

## Analizando los valores perdidos

**- PoolQC:** Según la descripción de los datos los *NA* significa que no tienen piscina, por eso hay tantos, así que vamos a cambiar los *NA* por *None*.

**- MiscFeature:** Los *NA* salen cuando no tienen ninguna característica miscelena.

**- Alley:** Los *NA* salen cuando la casa no tiene callejón de acceso.

**- Fence:** Los *NA* salen cuando la casa no tiene valla.

**- FireplaceQu:** Los *NA* salen cuando la casa no tiene chimenea.

Todos estos *NA* los vamos a cambiar por *None*:

```python
all_data["PoolQC"] = all_data["PoolQC"].fillna("None")
all_data["MiscFeature"] = all_data["MiscFeature"].fillna("None")
all_data["Alley"] = all_data["Alley"].fillna("None")
all_data["Fence"] = all_data["Fence"].fillna("None")
all_data["FireplaceQu"] = all_data["FireplaceQu"].fillna("None")
```

**- LotFrontage:** Dado que el área de cada calle conectada a la propiedad de la casa probablemente tenga un área similar a otras casas en su vecindario, podemos completar los valores faltantes por la mediana de *LotFrontage* del vecindario.

```python
#Agrupamos las ubicaciones físicas(barrios?) que vienen en Neighborhood y los agrupamos según los pies de cada calle conectada
#a la propiedad, luego hacemos la media de todo
#Group by neighborhood and fill in missing value by the median LotFrontage of all the neighborhood
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

**- GarageType, GarageFinish, GarageQual y GarageCond:** Reemplazamos los *NA* por *None*

```python
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    all_data[col] = all_data[col].fillna('None')
```

**- GarageYrBlt, GarageArea and GarageCars:** Reemplazamos los datos perdidos por `0`, deducimos que esos datos nos indican que no hay garaje.

```python
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    all_data[col] = all_data[col].fillna(0)
```

**- BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, BsmtFullBath and BsmtHalfBath:** Lo mismo que antes, de ducimos que tiene `0` sotanos.

```python
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    all_data[col] = all_data[col].fillna(0)
```

**- BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1 and BsmtFinType2:** Lo mismo que antes, de ducimos que no hay sotano.

```python
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    all_data[col] = all_data[col].fillna('None')
```

**- MasVnrArea and MasVnrType:** *NA* probablemente significa que no hay chapa de albañilería para estas casas. Podemos llenar `0` para el área y `None` para el tipo.

```python
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
```

**- MSZoning:** La clasificación general de zonificación, `RL` *(Residential Low Density)* es el valor más común. Entonces podemos completar los valores faltantes con `RL`.

```python
all_data['MSZoning'] = all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0]) #le decimos que lo sustituya por la moda (RL)
```

**- Utilities:** Para esta variable categórica que indica los servicios publicos que tiene la casa, todos los registros son `AllPub`, excepto uno `NoSeWa` y `2` `NA`. Dado que la casa con `NoSewa` está en el conjunto de entrenamiento, esta característica no ayudará en el modelado predictivo. Entonces podemos eliminarlo con seguridad.

```python
all_data = all_data.drop(['Utilities'], axis=1)
```

**- Functional:** En la descripción los *NA* son los típicos *(Typical)*.

```python
all_data["Functional"] = all_data["Functional"].fillna("Typ")
```

**- Electrical:** Cambiamos el único valor *NA* por la moda que es *SBrkr*.

```python
all_data['Electrical'] = all_data['Electrical'].fillna(all_data['Electrical'].mode()[0])
```

**- KitchenQual:** El mismo razonamiento que antes, lo cambiamos por su moda *TA*

```python
all_data['KitchenQual'] = all_data['KitchenQual'].fillna(all_data['KitchenQual'].mode()[0])
```

**- Exterior1st** y **Exterior2nd:** Lo mismo que arriba, al tener solo un valor perdido lo sustituimos por su moda.

```python
all_data['Exterior1st'] = all_data['Exterior1st'].fillna(all_data['Exterior1st'].mode()[0])
all_data['Exterior2nd'] = all_data['Exterior2nd'].fillna(all_data['Exterior2nd'].mode()[0])
```

**- SaleType:** Sustituimos el valor perdido por su moda *WD*.

```python
all_data['SaleType'] = all_data['SaleType'].fillna(all_data['SaleType'].mode()[0])
```

**- MSSubClass:** *NA* probablemente significa que no existe una clase de construcción. Podemos reemplazar los valores faltantes por `None`.

```python
all_data['MSSubClass'] = all_data['MSSubClass'].fillna("None")
```

Comprobamos que no hay ningún valor perdido que nos hayamos dejado:

```python
#Check remaining missing values if any 
all_data_na = (all_data.isnull().sum() / len(all_data)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data.head()
```

## Transformación de variables numéricas a categóricas.

```python
#MSSubClass=Identifica el tipo de vivienda involucrada en la venta con numeros
all_data['MSSubClass'] = all_data['MSSubClass'].apply(str)

#Changing OverallCond into a categorical variable. Califica el estado general de la casa con un numero del 1 al 10
all_data['OverallCond'] = all_data['OverallCond'].astype(str)

#Year and month sold are transformed into categorical features.
all_data['YrSold'] = all_data['YrSold'].astype(str)
all_data['MoSold'] = all_data['MoSold'].astype(str)
```

## LabelEncoder

Codifica etiquetas de una característica categórica en valores numéricos entre `0` y el número de clases menos `1`.

```python
from sklearn.preprocessing import LabelEncoder

cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold')

# process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[c].values)) 
    all_data[c] = lbl.transform(list(all_data[c].values))

# shape        
print('Shape all_data: {}'.format(all_data.shape))
```
