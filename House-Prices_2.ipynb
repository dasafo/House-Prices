## AnÃ¡lisis de SalePrice

```python
#histogram
sns.distplot(train['SalePrice'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

* Vemos que la grÃ¡fica superior no se ajusta a una normal y tiene un sesgo positivo.
* Viendo la grÃ¡fica inferior podemos ver que no sigue una relaciÃ³n lineal lo que nos indicarÃ­a que el residuo de sus errores no siguen una distribuciÃ³n normal.

Sabemos segÃºn la teorÃ­a que para distribuciones con un sesgo positivo podemos aplicarle una transformaciÃ³n logarÃ­tmica y obtendremos una relaciÃ³n lineal de los residuos de los errores:

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
train["SalePrice"] = np.log1p(train["SalePrice"])
```

```python
#Check the new distribution 
sns.distplot(train['SalePrice'] , fit=norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train['SalePrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train['SalePrice'], plot=plt)
plt.show()
```

Vemos que aplicando la transformaciÃ³n logaritmica tenemos una distribuciÃ³n normal.

## Analizando los valores perdidos

**- PoolQC:** SegÃºn la descripciÃ³n de los datos los *NA* significa que no tienen piscina, por eso hay tantos, asÃ­ que vamos a cambiar los *NA* por *None*.

**- MiscFeature:** Los *NA* salen cuando no tienen ninguna caracterÃ­stica miscelena.

**- Alley:** Los *NA* salen cuando la casa no tiene callejÃ³n de acceso.

**- Fence:** Los *NA* salen cuando la casa no tiene valla.

**- FireplaceQu:** Los *NA* salen cuando la casa no tiene chimenea.

Todos estos *NA* los vamos a cambiar por *None*:

```python
all_data["PoolQC"] = all_data["PoolQC"].fillna("None")
all_data["MiscFeature"] = all_data["MiscFeature"].fillna("None")
all_data["Alley"] = all_data["Alley"].fillna("None")
all_data["Fence"] = all_data["Fence"].fillna("None")
all_data["FireplaceQu"] = all_data["FireplaceQu"].fillna("None")
```

**- LotFrontage:** Dado que el Ã¡rea de cada calle conectada a la propiedad de la casa probablemente tenga un Ã¡rea similar a otras casas en su vecindario, podemos completar los valores faltantes por la mediana de *LotFrontage* del vecindario.

```python
#Agrupamos las ubicaciones fÃ­sicas(barrios?) que vienen en Neighborhood y los agrupamos segÃºn los pies de cada calle conectada
#a la propiedad, luego hacemos la media de todo
#Group by neighborhood and fill in missing value by the median LotFrontage of all the neighborhood
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
```

**- GarageType, GarageFinish, GarageQual y GarageCond:** Reemplazamos los *NA* por *None*

```python
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    all_data[col] = all_data[col].fillna('None')
```

**- GarageYrBlt, GarageArea and GarageCars:** Reemplazamos los datos perdidos por `0`, deducimos que esos datos nos indican que no hay garaje.

```python
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    all_data[col] = all_data[col].fillna(0)
```

**- BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, BsmtFullBath and BsmtHalfBath:** Lo mismo que antes, de ducimos que tiene `0` sotanos.

```python
for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    all_data[col] = all_data[col].fillna(0)
```

**- BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1 and BsmtFinType2:** Lo mismo que antes, de ducimos que no hay sotano.

```python
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    all_data[col] = all_data[col].fillna('None')
```

**- MasVnrArea and MasVnrType:** *NA* probablemente significa que no hay chapa de albaÃ±ilerÃ­a para estas casas. Podemos llenar `0` para el Ã¡rea y `None` para el tipo.

```python
all_data["MasVnrType"] = all_data["MasVnrType"].fillna("None")
all_data["MasVnrArea"] = all_data["MasVnrArea"].fillna(0)
```

**- MSZoning:** La clasificaciÃ³n general de zonificaciÃ³n, `RL` *(Residential Low Density)* es el valor mÃ¡s comÃºn. Entonces podemos completar los valores faltantes con `RL`.

```python
all_data['MSZoning'] = all_data['MSZoning'].fillna(all_data['MSZoning'].mode()[0]) #le decimos que lo sustituya por la moda (RL)
```

**- Utilities:** Para esta variable categÃ³rica que indica los servicios publicos que tiene la casa, todos los registros son `AllPub`, excepto uno `NoSeWa` y `2` `NA`. Dado que la casa con `NoSewa` estÃ¡ en el conjunto de entrenamiento, esta caracterÃ­stica no ayudarÃ¡ en el modelado predictivo. Entonces podemos eliminarlo con seguridad.

```python
all_data = all_data.drop(['Utilities'], axis=1)
```

**- Functional:** En la descripciÃ³n los *NA* son los tÃ­picos *(Typical)*.

```python
all_data["Functional"] = all_data["Functional"].fillna("Typ")
```

**- Electrical:** Cambiamos el Ãºnico valor *NA* por la moda que es *SBrkr*.

```python
all_data['Electrical'] = all_data['Electrical'].fillna(all_data['Electrical'].mode()[0])
```

**- KitchenQual:** El mismo razonamiento que antes, lo cambiamos por su moda *TA*

```python
all_data['KitchenQual'] = all_data['KitchenQual'].fillna(all_data['KitchenQual'].mode()[0])
```

**- Exterior1st** y **Exterior2nd:** Lo mismo que arriba, al tener solo un valor perdido lo sustituimos por su moda.

```python
all_data['Exterior1st'] = all_data['Exterior1st'].fillna(all_data['Exterior1st'].mode()[0])
all_data['Exterior2nd'] = all_data['Exterior2nd'].fillna(all_data['Exterior2nd'].mode()[0])
```

**- SaleType:** Sustituimos el valor perdido por su moda *WD*.

```python
all_data['SaleType'] = all_data['SaleType'].fillna(all_data['SaleType'].mode()[0])
```

**- MSSubClass:** *NA* probablemente significa que no existe una clase de construcciÃ³n. Podemos reemplazar los valores faltantes por `None`.

```python
all_data['MSSubClass'] = all_data['MSSubClass'].fillna("None")
```

Comprobamos que no hay ningÃºn valor perdido que nos hayamos dejado:

```python
#Check remaining missing values if any 
all_data_na = (all_data.isnull().sum() / len(all_data)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)
missing_data = pd.DataFrame({'Missing Ratio' :all_data_na})
missing_data.head()
```

## TransformaciÃ³n de variables numÃ©ricas a categÃ³ricas.

```python
#MSSubClass=Identifica el tipo de vivienda involucrada en la venta con numeros
all_data['MSSubClass'] = all_data['MSSubClass'].apply(str)

#Changing OverallCond into a categorical variable. Califica el estado general de la casa con un numero del 1 al 10
all_data['OverallCond'] = all_data['OverallCond'].astype(str)

#Year and month sold are transformed into categorical features.
all_data['YrSold'] = all_data['YrSold'].astype(str)
all_data['MoSold'] = all_data['MoSold'].astype(str)
```

## LabelEncoder

Codifica etiquetas de una caracterÃ­stica categÃ³rica en valores numÃ©ricos entre `0` y el nÃºmero de clases menos `1`.

```python
from sklearn.preprocessing import LabelEncoder

cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold')

# process columns, apply LabelEncoder to categorical features
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[c].values)) 
    all_data[c] = lbl.transform(list(all_data[c].values))

# shape        
print('Shape all_data: {}'.format(all_data.shape))
```
